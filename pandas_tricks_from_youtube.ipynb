```python
lists1 = pd.DataFrame({'col_one':['a','b','c'],
                       'col_two':[[10,40], [20,50], [30,60]]})
lists1
```

```python
lists2 = lists1.col_two.apply(pd.Series)
lists2
```

```python
lists3 = pd.concat([lists1, lists2], axis = 'columns')
lists3
```

```python
lists3 = lists3.drop('col_two', axis =1 )
```
